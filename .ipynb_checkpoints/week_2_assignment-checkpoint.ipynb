```python
import pandas as pd
import numpy as np

df = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv")
df.head()
```

```python
df
```

```python
categorical_data = df.select_dtypes(exclude=[np.number])
categorical_data.head()
```

```python
numeric_data = df.select_dtypes(include=[np.number])
numeric_data.head()
```

```python
numeric_data.describe()
```

i. Professors are more in numbers than assistant professors and associate professors.
ii. Discipline B is more than A.
iii. Males and Females are of equal numbers.

```python
%time
for i in categorical_data:
    val = df[i].value_counts()
    print()
    print("{} VALUE COUNTS: \n{}".format(i.upper(), val))
    print()
```

```python
# A Function to print out mean, sum and count of service, phd and salary
def count_mean(x):
    b = df.groupby(x, sort=False).agg({"service":["count", "sum", "mean"], "phd":["count", "sum", "mean"], "salary":["count", "sum", "mean"]})
    print()
    print("{}:".format(x.upper()))
    print(b)
```

```python
%time
for i in categorical_data:
    count_mean(i)
```

From the above, it is observed in RANK:
i.  Professors(Prof) dominate the rank than the others in both service mean, phd mean and salary mean.
ii. Professors(Prof) average a service of 21.41, which means number of service needed to be a Professors is high.
iii. Professors(Prof) average a PhD of 27.07, which means number of PhD needed to be a Professors is high.

From the above, it is observed in DISCIPLINE:
i.  Discipline A dominates the discipline category in both service mean and phd mean.
ii. Discipline B earns a higher salary at 116,331.79, than Discipline A at 98,331.11.

From the above, it is observed in SEX:
i.  MALE dominates the sex category in both service mean, phd mean and salary mean.
ii. MALES earn a higher salary at 115,045.15, than FEMALES at 101,002.41.

```python
# A Function to print out min and max of service, phd and salary
def min_max(x):
    b = df.groupby(x, sort=False).agg({"service":["min", "max"], "phd":["min", "max"], "salary":["min", "max"]})
    print()
    print("{}:".format(x.upper()))
    print(b)
```

```python
%time
for i in categorical_data:
    min_max(i)
```

The table above shows the minimum and maximum values of service, PhD and salary for each categorical data.
